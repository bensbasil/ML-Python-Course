```python
##Merging and joining Data Frames
df1=pd.DataFrame({'Key':['A','B','C'],'Value1':[1,2,3]})
df2=pd.DataFrame({'Key':['A','B','D'],'Value1':[4,5,6]})
```

```python
df1
```

```python
df2
```

```python
##Merge DataFrame on the 'Key Columns'
df_merged=pd.merge(df1,df2,on="Key",how="inner")
df_merged
```

```python
df_merged=pd.merge(df1,df2,on="Key",how="outer")
df_merged
```

```python
df_merged=pd.merge(df1,df2,on="Key",how="left")
df_merged
```

```python
df_merged=pd.merge(df1,df2,on="Key",how="right")
df_merged
```
